## ðŸ“Œ Importing Necessary Libraries for Model Fine-Tuning and Dataset Handling








```python
import os
import unsloth
```

```python
from datasets import load_dataset, Dataset
import torch
```

```python
from huggingface_hub import notebook_login
notebook_login()
```

```python
from unsloth import FastVisionModel
from datasets import load_dataset
from transformers import TextStreamer
from unsloth import is_bf16_supported
from unsloth.trainer import UnslothVisionDataCollator
from trl import SFTTrainer, SFTConfig
from unsloth import FastLanguageModel
from trl import SFTTrainer
from transformers import TrainingArguments, AutoModelForCausalLM, AutoTokenizer
```

## ðŸ“Œ Loading and Preprocessing the Radiology Dataset for Fine-Tuning


```python

dataset = load_dataset("unsloth/Radiology_mini", split = "train")
instruction = "You are an expert radiographer. Describe accurately what you see in this image."

def convert_to_conversation(sample):
    conversation = [
        { "role": "user",
          "content" : [
            {"type" : "text",  "text"  : instruction},
            {"type" : "image", "image" : sample["image"]} ]
        },
        { "role" : "assistant",
          "content" : [
            {"type" : "text",  "text"  : sample["caption"]} ]
        },
    ]
    return { "messages" : conversation }
pass

converted_dataset = [convert_to_conversation(sample) for sample in dataset]
```

## ðŸ“Œ Loading the Pretrained LLaMA 3.2 Vision Model for Fine-Tuning


```python
model,tokenizer = FastVisionModel.from_pretrained(
    model_name = 'unsloth/Llama-3.2-11B-Vision-Instruct',
    max_seq_length = 2048,
    dtype = None,
    load_in_4bit = False)
```

```python
print(model)
print(tokenizer)
```

## ðŸ“Œ Applying LoRA for Parameter-Efficient Fine-Tuning of the Vision Model


```python
model_unsloth = FastVisionModel.get_peft_model(
    model,
    r=16,
    lora_alpha=16,
    lora_dropout=0,
    target_modules=['q_proj',"k_proj","v_proj","o_proj","gate_proj",'up_proj',"down_proj"],
    bias="none",
    use_gradient_checkpointing=True,
    random_state=42,
    max_seq_length=2048
)
```

## ðŸ“Œ Configuring LoRA Fine-Tuning Parameters for Optimized Training


```python
 args = SFTConfig(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 8,
        warmup_steps = 5,
        max_steps = 10,
        learning_rate = 2e-4,
        bf16 = is_bf16_supported(),
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "cosine",
        seed = 3407,
        output_dir = "content/unsloth_llama3.2",
        report_to = "none",
        remove_unused_columns = False,
        dataset_text_field = "",
        dataset_kwargs = {"skip_prepare_dataset": True},
        dataset_num_proc = 4,
        max_seq_length = 2048,
    )
```

## ðŸ“Œ Checking GPU Specifications and Memory Usage for Training


```python
gpu_stats = torch.cuda.get_device_properties(0)
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
max_memory = round(gpu_stats.total_memory / 1024 / 1024 / 1024, 3)
print(f"GPU = {gpu_stats.name}. Max memory = {max_memory} GB.")
print(f"{start_gpu_memory} GB of memory reserved.")
```

## ðŸ“Œ Initializing and Running the Fine-Tuning Trainer for Vision Model


```python
trainer = SFTTrainer(
    model = model_unsloth,
    tokenizer = tokenizer,
    data_collator = UnslothVisionDataCollator(model, tokenizer),
    train_dataset = converted_dataset,
    max_seq_length = 2048,
    args = args,
    )
```

```python
stats = trainer.train()
```

## ðŸ“Œ Performing Inference with the Fine-Tuned Vision Model for Radiology Analysis


```python

FastVisionModel.for_inference(model_unsloth)
image = dataset[0]["image"]
instruction = "You are an expert radiographer. Describe accurately what you see in this image."

messages = [
    {"role": "user", "content": [
        {"type": "image"},
        {"type": "text", "text": instruction}
    ]}
]
input_text = tokenizer.apply_chat_template(messages, add_generation_prompt = True)
inputs = tokenizer(
    image,
    input_text,
    add_special_tokens = False,
    return_tensors = "pt",
).to("cuda")

text_streamer = TextStreamer(tokenizer, skip_prompt = True)
_ = model.generate(**inputs, streamer = text_streamer, max_new_tokens = 128,
                   use_cache = True, temperature = 1.5, min_p = 0.1)
```

## ðŸ“Œ Saving and Uploading the Fine-Tuned Vision Model to Hugging Face Hub


```python
model_unsloth.save_pretrained("lora_model")
tokenizer.save_pretrained("lora_model")

model_unsloth.save_pretrained_merged("Varu96/llama-3.2-11B-Vision-Medical", tokenizer,)
model_unsloth.push_to_hub_merged("Varu96/llama-3.2-11B-Vision-Medical", tokenizer, save_method = "merged_16bit", token = os.environ.get("You"))
```
